VISUALIZATION STRESS DISTRIBUTION

```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```
def f(x, y):
    return 100*(np.sqrt(5/(2*y)))*(np.cos(x/2)*(1-np.sin(x/2)*np.sin(3*x/2)))

def f1(x, y):
    return 100*(np.sqrt(5/(2*y)))*(np.cos(x/2)*(1+np.sin(x/2)*np.sin(3*x/2)))

def f2(x, y):
    return 100*(np.sqrt(5/(2*y)))*(np.cos(x/2)*np.sin(x/2)*np.cos(3*x/2))

x = np.linspace(0, 360, 360) # angle
y = np.linspace(0.1, 10, 360) # radius

X, Y = np.meshgrid(x, y)

Z = f(X, Y) # sigma_x
Z1 = f1(X, Y) # sigma_y
Z2 = f2(X, Y) # tao_xy
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.set_title('Variation of sigma_x')
```

```
# To rotate the plot
ax.view_init(6, 50)
fig 
```

```
fig1 = plt.figure()
ax1 = plt.axes(projection='3d')
ax1.contour3D(X, Y, Z1, 50, cmap='viridis')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z');
ax1.set_title('Variation of sigma_y')
```

```
# To rotate the plot
ax.view_init(6, 50)
fig
```

```
fig2 = plt.figure()
ax2 = plt.axes(projection='3d')
ax2.contour3D(X, Y, Z2, 50, cmap='viridis')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z');
ax2.set_title('Variation of tau_xy')
```

```
# To rotate the plot
ax.view_init(6, 50)
fig
```
